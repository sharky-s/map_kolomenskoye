```python
import geopandas as gpd
import folium
```

```python
# Чтение GeoJSON файла
geojson_file = "path.geojson"
gdf = gpd.read_file(geojson_file)
```

```python
print(gdf.head())
```

```python
# Создание карты
m = folium.Map(location=[55.65, 37.66], zoom_start=14)
```

```python
# Добавление пути на карту
folium.GeoJson(gdf, name="Path").add_to(m)
```

```python
# Добавление точек интереса на карту
for idx, row in gdf.iterrows():
    if row.geometry.type == "Point":
        folium.Marker(
            location=[row.geometry.y, row.geometry.x],
            popup=folium.Popup("<b>" + row['iconCaption'] + "</b>" + "<br>" + "<p>" + row['description'] + "</p>" +
                            "<img src='" + row['image'] + "' style='width: 400px; height: 100%;'>"),
            icon=folium.Icon(color=row['marker-color'])
        ).add_to(m)
```

```python
m.save("map.html")
```
